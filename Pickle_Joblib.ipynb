```python
import pandas as pd
from sklearn import linear_model
data = pd.read_csv("/content/canada_per_capita_income.csv")


data
x = data[['year']]
y = data[['per capita income (US$)']]
model = linear_model.LinearRegression()
model.fit(x,y)
print(model.predict([[3000]]))




import pickle
with open ('model_picle_test','wb') as file:
  pickle.dump(model,file)
with open ('model_picle_test','rb') as file:
  m3 = pickle.load(file)

print(m3.predict([[3000]]))

```

```python
import joblib
joblib.dump(model,'model_joblib_test')
m4 = joblib.load('model_joblib_test')
print(m4.predict([[3000]]))
```
